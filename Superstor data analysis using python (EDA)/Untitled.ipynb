# Sales Trend analysis

```python
df.head()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'],dayfirst=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df_ys=df.groupby(df['Order Date'].dt.year)['Sales'].agg(['sum','mean','count']).reset_index()
df_ys.sort_values(by='sum',ascending=False)
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df_ys,x='Order Date',y='sum')
plt.title('Sum of sales by year')
plt.xlabel('Year')
plt.ylabel('Sum of sales')
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.plot(df_ys['Order Date'],df_ys['sum'],marker='o',linestyle='--')
plt.title('Total sales by year')
plt.xlabel('Year')
plt.ylabel('Sum of sales')
plt.xticks(rotation=65)
plt.show()
```

# Quartely Sales

```python
df.head()
```

```python
Sales_2018=df[df['Order Date'].dt.year==2018]
Sales_2018.head()
```

```python
quartely_sales=Sales_2018.resample('Q',on='Order Date')['Sales'].sum()
```

```python
quartely_sales=quartely_sales.reset_index()
quartely_sales=quartely_sales.rename(columns={'Order Date':'Quarter','Sales':'Total Sales'})
quartely_sales
```

```python
plt.figure(figsize=(8,4))
plt.plot(quartely_sales['Quarter'],quartely_sales['Total Sales'],marker='o',linestyle='--')
plt.title('Total sales by Quarter')
plt.xlabel('Quarters')
plt.ylabel('Sum of sales')
plt.xticks(rotation=65)
plt.show()
```

# Trends by Month

```python
df_m=df.groupby(df['Order Date'].dt.month)['Sales'].sum().reset_index()
df_m
```
